```python
import pandas as pd

# read csv file
df1=pd.read_csv("dog_list.csv")

#get the length of file 
len(df1)

#get random number
import random
data=random.randint(0,196)
print(data)

random_dog = df1.iloc[data]["Breed names"]
print(random_dog)

#get URL
url= "https://dog.ceo/api/breed/{}/images/random".format(random_dog)
print(url)

from datetime import datetime
date = datetime.utcnow()
file_name = f"dog_{random_dog}_{date.strftime('%Y-%m-%d')}.jpg"
print(file_name)
```
